```python
%%writefile multipleblocks.cu

#include <stdio.h>

// Kernel function where each thread prints its global thread ID
__global__ void helloWorldKernel()
{
    // Calculate the global thread ID
    int globalThreadId = blockIdx.x * blockDim.x + threadIdx.x;

    // Print "Hello World" along with the global thread ID
    printf("Hello World from thread %d (Block %d, Thread %d)\n",
           globalThreadId, blockIdx.x, threadIdx.x);
}

int main()
{
    // Define the number of blocks and threads per block
    int numBlocks = 3;
    int threadsPerBlock = 5;

    // Launch the kernel with 'numBlocks' blocks and 'threadsPerBlock' threads per block
    helloWorldKernel<<<numBlocks, threadsPerBlock>>>();

    // Wait for all threads to complete and synchronize
    cudaDeviceSynchronize();

    return 0;
}
```

```python
!nvcc -o multipleblocks multipleblocks.cu
```
