```python
%%writefile helloworld.cu

#include <stdio.h>

// Kernel function where each thread prints its thread ID
__global__ void helloWorldKernel()
{
    // Get the thread ID
    int threadId = threadIdx.x;

    // Print "Hello World" along with the thread ID
    printf("Hello World from thread %d\n", threadId);
}

int main()
{
    // Define the number of threads per block
    int numThreads = 10;

    // Launch the kernel with one block and 'numThreads' threads
    helloWorldKernel<<<1, numThreads>>>();

    // Wait for all threads to complete and synchronize
    cudaDeviceSynchronize();

    return 0;
}
```

```python
!nvcc helloworld.cu -o helloworld
```
